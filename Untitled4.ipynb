```python
from html.parser import HTMLParser
import requests


def _pretty_print(x, preffix=""):
    if isinstance(x, list) or isinstance(x, tuple) or isinstance(x, set):
        return preffix + "[\n" + u",\n".join([_pretty_print(y, preffix + "    ") for y in x]) + "\n" + preffix + "]"
    elif isinstance(x, dict):
        return (preffix + "{\n" + 
                ",\n".join([u"{}: {}".format(
                    _pretty_print(a, preffix + "    "), _pretty_print(b, preffix + "    ").strip()) 
                            for a, b in sorted(x.items())]
                          ) + "\n" + preffix + "}")
    else:
        return preffix + str(x)   

def pretty_print(x):
    print (_pretty_print(x))


class MyParser(HTMLParser):    
    def __init__(self):
        HTMLParser.__init__(self)
        self.result = []
        self.possible_titles = [u"Ученик/Цель", u"Район", u"Формат", u"Ставка"]
        self.myreset()
        
        
    def myreset(self):
        self.cur_title = None
        self.cur_row = {}   
        self.in_orders = False
        self.got_empty_div = 0

        
    def handle_starttag(self, tag, attrs):
        attrs_dict = dict(attrs)
        
        if tag == "table":
            if attrs_dict.get("class") == "order_block ":
                self.in_orders = True
        if self.in_orders:
            if tag == "a":
                if u"Имя" in self.cur_row:
                    self.cur_title = None
                else:
                    self.cur_title = u"Имя"
            if tag == "div":
                self.cur_title = attrs_dict.get("title")
                if not self.cur_title:
                    if self.got_empty_div == 0:
                        self.cur_title = u"Предмет"
                    if self.got_empty_div == 1:
                        self.cur_title = u"Пол преподавателя"
                    self.got_empty_div += 1
                else:
                    if self.cur_title not in self.possible_titles:
                        self.cur_title = None
        if self.cur_title and self.cur_title not in self.cur_row:
            self.cur_row[self.cur_title] = []

            
    def handle_endtag(self, tag):
        if tag == "table" and self.in_orders:
            self.cur_row[u"Имя"] = self.cur_row[u"Имя"][1]
            if u"Пол преподавателя" not in self.cur_row:
                self.cur_row[u"Пол преподавателя"] = u"Любой"
            elif self.cur_row[u"Пол преподавателя"]:
                self.cur_row[u"Пол преподавателя"] = self.cur_row[u"Пол преподавателя"][1]
            else:
                self.cur_row[u"Пол преподавателя"] = u"Любой"
            self.cur_row[u"Предмет"] = self.cur_row[u"Предмет"][0]
            self.result.append(self.cur_row)
            self.myreset()

            
    def handle_data(self, data):
        if self.in_orders:
            if self.cur_title and data.strip():
                if self.cur_title == u"Пол преподавателя":
                    if data.strip() != u"Пол преподавателя =" and not self.cur_row[self.cur_title]:
                        self.cur_title = None
                        return
                self.cur_row[self.cur_title].append(data.strip())
```

```python
r = requests.get('https://repetitors.info/backoffice/n.php?all=1', auth=('PavlovVK', 'LllTraZDx3'))
```

```python
parser = MyParser()
parser.feed(r.text)
result = parser.result
```

```python
def res(cur_str):
    if '…' in cur_str:
        index = cur_str.find('…')
        return cur_str[:index]
    else:
        return cur_str
```

```python
import csv

list_learner1 = ['11 кл.', '10 кл.', '9 кл.']
list_learner2 = ['8 кл.', '7 кл.', '6 кл.', '5 кл.']
list_learner3 = ['4 кл.', '3 кл.', '2 кл.', '1 кл.']
list_learner4 = ['дошкольник', 'ребёнок']

with open('profi_ru.csv', "w", newline="") as file:
    columns = ["Имя", "Предмет", "Пол преподавателя", "Район", "Ставка", "Продолжительность", "Ученик"]
    writer = csv.DictWriter(file, fieldnames=columns)
    writer.writeheader()
    for el in result:
        name = el['Имя']
        theme = el['Предмет'].split(';')[0]
        gender = el['Пол преподавателя']
        area = el.get('Район', None)
        area = area[0].split(',')[0] if area else area
        rate = res(el['Ставка'][0].split(' ')[-1])
        duration = el['Ставка'][-1].split(' ')[0]
        learner = el['Ученик/Цель'][0]
        if learner in list_learner1:
            learner = '9 - 11'
        elif learner in list_learner2:
            learner = '5 - 9'
        elif learner in list_learner3:
            learner = '1 - 4'
        elif learner in list_learner4:
            learner = 'дошкольник'
        elif learner == 'взрослый':
            learner = 'взрослый'
        elif learner == 'студент':
            learner = 'студент'
        else:
            learner = None
        user = {"Имя": name,
            "Предмет": theme,
            "Пол преподавателя": gender,
            "Район": area,
            "Ставка": rate,
            "Продолжительность": duration,
            "Ученик": learner
        }
        writer.writerow(user)
```
